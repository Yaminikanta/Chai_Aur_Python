```python
"""
above example gives result but always perform operation with same data type

we can explicitly convert data type by using typecasting 

Typycasting just changes the data type

"""

a = 40
b = float(a)        # changed to float type
print('a-',a)
print('b-',b)

c =20.23
d = int(c)          # changed to int
print('c-',c)
print('d-',d)

print(40 + int(20.22))
print(float(40) + 20.22)
```
