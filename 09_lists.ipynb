### Lists

```python
tea_varieties = ['black', 'green', 'oolang', 'white']
print(tea_varieties)
print(tea_varieties[0])
print(tea_varieties[1])
print(tea_varieties[-1])
print(tea_varieties[:])
```

```python
# list is mutable so we can change value
tea_varieties[3] = 'herbal'
print(tea_varieties)
```

```python
for tea in tea_varieties:
    print(tea)
```

```python
for tea in tea_varieties:
    print(tea, end= '-')
```

```python
if 'green' in tea_varieties:
    print('I have green tea')
else:
    print('I have not green tea')
```

```python
if 'masala' in tea_varieties:
    print('I have masala tea')
else:
    print('I have not masala tea')
```

```python
tea_varieties.append('masala')
print(tea_varieties)
```

```python
if 'masala' in tea_varieties:
    print('I have masala tea')
else:
    print('I have not get masala tea')
```
