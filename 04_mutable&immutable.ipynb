```python
username = "yamini kant"
print(username)
print(id(username), '\n')

username = 'chai aur code'
print(username)
print(id(username))
```

```python
x = 10
y = x 
print('x = ', x, ', id-', id(x))
print('y = ', y, ', id-', id(y),'\n') 

x = 15
print('x = ', x, ', id-', id(x))
print('y = ', y, ', id-', id(y)) 
```
