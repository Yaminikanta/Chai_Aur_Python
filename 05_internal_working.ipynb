```python
"""
-> In it we have taken immutable data type i.e int float and string 
-> In immutable data type when we assign a new value to same variable it creates a new memory space for it and the old data or 
   value will go to Garbage collection 

-> In python when we reassign a new value to same variable it pushes the old value to Garbage collection and remembers the latest value

-> In the below example we see how immuttable data type works, every time when value is changed and because the data type is immutable
   so every time it creating a new space(memory id)
   
"""

a = 3
a = 'chai aur code'
a = 3.14
a = 5
b = 2

print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b))

# even when we add value it will also change memory id
a = a+2
print('a = ',a, ', id-', id(a))
```

```python
a = 1234
print('a = ',a, ', id-', id(a))

a = 1234
print('a = ',a, ', id-', id(a),'(even wheater value is same still it allocates new memory space  because of immutable in nature)')

a = 'chai'
print('a = ',a, ', id-', id(a))

a = 3.14
print('a = ',a, ', id-', id(a))

a = 5
b = 2
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b))

a = a+2
print('a = ',a, ', id-', id(a))
```

```python
"""
-> Now we will understand how mutuable data type works 
-> In the below example we see how muttable data type works, every time when value is changed and because the data type is mutable
   so every time it will not creating a new memory space it will just change the data 

"""
myListOne = [1, 2, 3]
myListTwo = myListOne
print(myListOne, ', id-',  id(myListOne))             # in it both listOne and listTwo have same refference id
print(myListTwo, ', id-',  id(myListTwo),'\n')        # in it both listOne and listTwo have same refference id 

myListOne = 'chai'                                    # here we reassign the value to 'chai' (which is a immutable data type)
print(myListOne, ', id-',  id(myListOne))             # here we will get a new id bcz we reassign with a new value
print(myListTwo, ', id-',  id(myListTwo),'\n')        # listTwo id is same as previous

myListOne = [1, 2, 3]                                 # here may be value looks same but we reassign it so memory refference will be different
print(myListOne, ', id-',  id(myListOne))             # got a new memory refference
print(myListTwo, ', id-',  id(myListTwo),'\n')        # listTwo id is same as previous

myListOne[0] = 143                                    # here we change the first index value (as list is mutuable)
print(myListOne, ', id-',  id(myListOne))             # but the memory reference is same as previous reassigned
print(myListTwo, ', id-',  id(myListTwo))             # listTwo id is same as previous
```

## Copy Operation

```python
"""
->A copy operation in Python means creating a new variable or object based on an existing one so that data can be reused without
(or with limited) shared memory.

-> It is of 3 types
1) normal copy
2) shallow copy
3) deep copy

"""
```

```python
"""
Normal copy:- 
In it we will see how Normal copy works using id, its drop back and how to over come it

In it we see when we copy a value from source variable(a) to destination variable(b) along with value the reference id also copied
and if we change anything in any variable the other variable also got changed which is not good 
means here Only reference is copied and Changes reflect in both variables

to over come this problem we will use shallow copy

"""
a = [1, 2, 3]
b = a
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

a[0] = 11
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

b[1] = 22
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b))
```

```python
"""
Shallow copy:-
In it A new outer object is created, but inner objects are shared.

to use it we have to import copy module or we can use slice

in the below example we can see both a and b have different ref id with same value and when we change value using indexing to does not 
impact other variable

"""
# using copy module

import copy

a = [1, 2, 3]
b = copy.copy(a)
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

a[0] = 11
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

b[1] = 22
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b))
```

```python
# using slice

a = [1, 2, 3, 4, 5]
b = a[:]
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

a[0] = 10
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

b[1] = 20
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b))
```

```python
"""
but there is a problem in shallow copy when we work with nested data when values are same it got same reference id
and bcz of same reference id and its mutable so when we change value in 'a' it reflects with 'b' also and vice versa

so to over come this problem we will use deep copy in next step
"""
import copy

a = [[1, 2, 3], [4, 5, 6]]
b = copy.copy(a)
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')      

print('a[0] = ', a[0], ', id-', id(a[0]),  ', a[1] = ', a[1], id(a[1]))
print('b[0] = ', b[0], ', id-', id(b[0]),  ', b[1] = ', b[1], id(b[1]), '\n')

a[0][0] = 10
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

b[1][0] = 40
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

print('a[0] = ', a[0], ', id-', id(a[0]),  ', a[1] = ', a[1], id(a[1]))
print('b[0] = ', b[0], ', id-', id(b[0]),  ', b[1] = ', b[1], id(b[1]))
```

```python
"""
Deep Copy:-
In it all nested value along with whole variable also got different memory space so,
when we change any value in any position it does not impact to other variable

"""
import copy

a = [[1, 2, 3, 4], [5, 6, 7, 8]]
b = copy.deepcopy(a)

print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')      

print('a[0] = ', a[0], ', id-', id(a[0]),  ', a[1] = ', a[1], id(a[1]))
print('b[0] = ', b[0], ', id-', id(b[0]),  ', b[1] = ', b[1], id(b[1]), '\n')

a[0][0] = 10
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

b[1][0] = 50
print('a = ',a, ', id-', id(a))
print('b = ',b, ', id-', id(b),'\n')

print('a[0] = ', a[0], ', id-', id(a[0]),  ', a[1] = ', a[1], id(a[1]))
print('b[0] = ', b[0], ', id-', id(b[0]),  ', b[1] = ', b[1], id(b[1]))
```
