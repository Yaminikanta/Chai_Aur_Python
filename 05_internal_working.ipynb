```python
# In this example we will clarify how python memory allocation or memory refference works with the help of id()
# also copy operation, mutuable & immutable works. 

myListOne = [1, 2, 3]
myListTwo = myListOne
print(myListOne, id(myListOne))                 # in it both listOne and listTwo have same refference id
print(myListTwo, id(myListTwo),'\n')            # ini t both listOne and listTwo have same refference id

myListOne = 'chai'                              # here we reassign the value to 'chai'
print(myListOne, id(myListOne))                 # here we will get a new id bcz we reassign with a new value
print(myListTwo, id(myListTwo),'\n')            # listTwo id is same as previous

myListOne = [1, 2, 3]                           # here may be value looks same but we reassign it so memory refference will be different
print(myListOne, id(myListOne))                 # got a new memory refference
print(myListTwo, id(myListTwo),'\n')            # listTwo id is same as previous

myListOne[0] = 143                              # here we change the first index value (as list is mutuable)
print(myListOne, id(myListOne))                 # but the memory reference is same as previous reassigned
print(myListTwo, id(myListTwo))                 # listTwo id is same as previous
```

```python
l1 =[1, 2, 3]
l2 = l1
print(l1, id(l1))
print(l2, id(l2), '\n')

l1[0] = 10
print(l1, id(l1))
print(l2, id(l2), '\n')

l2[0] = 100
print(l1, id(l1))
print(l2, id(l2))
```
